```python
import sys
import os
import pandas as pd

# Ajout du dossier racine du projet au PYTHONPATH (pour trouver le module src)
sys.path.append(os.path.abspath('..'))

# Import de ta fonction
from src.anonymize import anonymize_comments
```

## 1. Test sur des phrases simples

```python
sample = (
    "Allison Hill is a strong promotion candidate this year. "
    "Discussed with HR (Rhonda Smith, 651.216.1559). "
    "Budget pre-approved on account ****3503."
)

print("Avant :\n", sample)
print("\nAprès :\n", anonymize_comments(sample, random_seed=42))
```

## 2. Test sur un DataFrame (simulation de CSV)

```python
# Simulation d'un petit jeu de données brut
data = {
    "id": [1, 2],
    "comment": [
        "John Doe and Jane Smith performed well. Contact: john.doe@company.com or jane.smith@mail.fr. Phone: 555-123-4567.",
        "Excellent work from Michael on the Paris project. No issues raised."
    ]
}

df = pd.DataFrame(data)
display(df)
```

```python
# Création d'une nouvelle colonne anonymisée pour pouvoir comparer
df["comment_anonymized"] = df["comment"].apply(lambda x: anonymize_comments(x, random_seed=42))
display(df)
```

## 3. Appliquer l'anonymisation sur un VRAI fichier CSV
Le code ci-dessous te montre la structure typique pour ouvrir ton vrai fichier `.csv`, l'anonymiser et le sauvegarder. *(Décommente la cellule pour t'en servir)*.

```python
input_file = "../data/audit_sample.csv"
output_file = "../data/audit_sample_anonymized_notebook.csv"
nom_colonne = "manager_comments"

# 1. Lire le CSV
df_reel = pd.read_csv(input_file)

# 2. Anonymiser (en traitant les valeures vides NaN possibles)
df_reel[nom_colonne] = df_reel[nom_colonne].fillna("").apply(anonymize_comments)

# 3. Écrire le CSV
df_reel.to_csv(output_file, index=False)
print("Anonymisation terminée et fichier sauvegardé !")
```
